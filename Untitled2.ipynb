#### Intro to slicing arrays

If we have an array called a_array and it has 5 elements with a size of 6, [-1, 3, 0, 6, 20, 8.4]




```python
a_array = [-1, 2, 0, 6, 20, 8.4]
## Slicing format of a[n:m] where m is not included
a_array[2:5] ##This will print elements 2 to 4 not include 5 [0, 6, 20]
## The colon (:) is the slice operator which cuts up the desired amount of elements in an array
a_array[4:] ##This will print elements 4 till the end of the array [20, 8.4]
a_array[:2] ##This will include all elements from 0 to n - 1.
## In this case since the last element is 2 it will include 0 to 1 elements
a_array[2:-1] ## This is a bit interesting!
## Here we start at index 2, but then we have -1 this will not give a syntax error
## What -1 is doing is starting at the end of the array from 0 (8.4) to -1 (20)
```

What if we have more than 1 dimension?
let's say we have a second array called b_array:
       
$\begin{bmatrix}
   2 & 4 & 8 \\
   2 & -1 & 0 \\
   3.5 & 28 & 32 
\end{bmatrix}$



```python
b_array = np.array([[2, 4, 8], [2, -1, 0], [3.5, 28, 32]])

b_array[2][0] ##i is the first brackets with rows counting from 0 to n
## We are selecting row 2, j the the second brackts which is the column
## We are selecting row 2, at column 0 which is 3.5

b_array[0] ## Now we are only taking all elements in row 0
b_array[:, 1]## What if we just want the columns? 
## We just need to slice the rows and take all elements from row 2
```

```python
## Lets make more data, but with a LOT more random data points!
## Instead of using rand we are usinf randn
## randn returns data points to Gaussian distribution
moreData = np.random.randn(100000,3)

x_array = moreData[:,0]
y_array = moreData[:,1]
z_array = moreData[:,2]

plt.plot(x_array, z_array, 'k.')
plt.title("Random Distribution")
```

```python
## Lets make a 2D histogram plot with the distribution
## the bin parameter increases the resultion the higher the bins
## this give more clear data to work with 
plt.hist2d(x_array, y_array, bins = 100)
plt.colorbar() ## Adds a color bar
plt.title("Random Distribution of points") ## adds title
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
